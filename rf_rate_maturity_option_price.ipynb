```python
import pandas as pd
from datetime import datetime
```

## Get Risk Free rates from daily treasury rates

```python
treasury_rates = pd.read_csv('daily-treasury-rates.csv', index_col='Date')
treasury_rates
```

```python
#selected_dates = treasury_rates.loc[['07/26/2024','08/30/2024','09/27/2024']]
selected_dates = treasury_rates.loc[['09/30/2024']]
selected_dates
```

```python
df4 = pd.read_csv('data/SPX/UnderlyingOptionsIntervals_3600sec_calcs_oi_2024-10-11.csv')
df5 = pd.read_csv('data/SPX/UnderlyingOptionsIntervals_3600sec_calcs_oi_2024-10-18.csv')
df6 = pd.read_csv('data/SPX/UnderlyingOptionsIntervals_3600sec_calcs_oi_2024-10-25.csv')
```

```python
df4.head()
```

```python
def days_between_dates(date_str1, date_str2):
    # Convert strings to datetime objects
    date1 = datetime.strptime(date_str1, "%Y-%m-%d")
    date2 = datetime.strptime(date_str2, "%Y-%m-%d")
    
    # Calculate the difference in days
    delta = abs((date2 - date1).days)
    
    return delta
```

```python
def get_period(days):
    if days <= 30:
        return "1 Mo"
    elif days <= 60:
        return "2 Mo"
    elif days <= 90:
        return "3 Mo"
    elif days <= 120:
        return "4 Mo"
    elif days <= 180:
        return "6 Mo"
    elif days <= 365:
        return "1 Yr"
    elif days <= 730:
        return "2 Yr"
    elif days <= 1095:
        return "3 Yr"
    elif days <= 1825:
        return "5 Yr"
    elif days <= 2555:
        return "7 Yr"
    else:
        return "10 Yr"
```

## For Each Df, add rf_rate, time to maturity (days / 252), option prices (mid point of ask and bid)

```python
df1_date = "2024-10-11"
treasury_rate_date = "09/30/2024"

rf_rates = []
time_to_maturity = []
option_prices = []

for index, row in df4.iterrows():
    expiration = row['expiration']
    maturity = days_between_dates(df1_date,expiration)
    time_to_maturity.append(maturity/252) #time to maturity in days/year
    rf_rate_period = get_period(maturity)
    rf_rates.append(selected_dates.loc[treasury_rate_date,rf_rate_period])
    option_price = (row['ask'] + row['bid'])/2
    option_prices.append(option_price)
    
```

```python
df4['risk_free_rate'] = rf_rates
```

```python
df4['time_to_maturity'] = time_to_maturity
```

```python
df4['option_price'] = option_prices
```
